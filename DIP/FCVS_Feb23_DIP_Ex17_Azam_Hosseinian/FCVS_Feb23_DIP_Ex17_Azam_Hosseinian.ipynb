### Use edge detection methods, to find the watch numbers' edges, you can use denoising methods before detecting edges!. ( the output has to be so clear, without any noise!)

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_img_plt(img, c_map ='gray', fig_hight=8, fig_width=6, vmin=0, vmax=255):
    
    plt.figure(figsize=(fig_hight, fig_width), dpi=100)
    plt.imshow(img, cmap=c_map, vmin=vmin, vmax=vmax)
```

```python
def show_mult_img(rows, columns, img_names, vmin=0, vmax=255):
    
    fig = plt.figure(figsize=(15, 17), dpi=100)
    for i in range(len(img_names)):
        fig.add_subplot(rows, columns, i+1)
        plt.imshow(img_names[i], cmap='gray', vmin=vmin, vmax=vmax)
        plt.axis('off')
        plt.title('img_' + str(i))
```

```python
#read images
input_img = cv2.imread('./img/input.jpg')
output_img = cv2.imread('./img/output.jpg')

show_mult_img(1, 2, (input_img, output_img))
```

```python
img = cv2.imread('./img/input.jpg')
show_img_plt(img)

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_blur = cv2.GaussianBlur(img_gray, (3, 3), 0)

show_img_plt(img_blur)
```

## Laplacian Edge Detection

```python
laplacian = cv2.Laplacian(img_blur, -1)
threshold_value = 25
ret, thresh = cv2.threshold(laplacian, threshold_value, 255, cv2.THRESH_BINARY)
show_mult_img(1, 2, (output_img, thresh))
```

## Canny Edge Detector

```python
canny = cv2.Canny(image=img_blur, threshold1=200, threshold2=250)
show_mult_img(1,2,(output_img, canny))
```

canny better than Laplacian
